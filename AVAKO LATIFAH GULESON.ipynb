```python


#AVAKO LATIFAH GULESON
#21/1/324/D/240 

import matplotlib.pyplot as plt

def mean(x, y):
    sum_x = sum(x)
    sum_y = sum(y)
    n = len(x)
    return sum_x/n, sum_y/n

def standard_dev(x, y):
    mean_x, mean_y = mean(x, y)
    sum1 = 0
    for i in range(len(x)):
        sum1 += ((x[i] - mean_x)**2 + (y[i] - mean_y)**2)
    return ((sum1/len(x)) ** 0.5)

def variance(x, y):
    mean_x, mean_y = mean(x, y)
    sum1 = 0
    for i in range(len(x)):
        sum1 += ((x[i] - mean_x)**2 + (y[i] - mean_y)**2)
    return sum1/len(x)

# Define the coordinates of the polygon
x = [10, 11, 12, 13, 14, 15]
y = [1, 4, 6, 5, 3, 2]

# Compute the mean, standard deviation, and variance
mean_x, mean_y = mean(x, y)
sd = standard_dev(x, y)
var = variance(x, y)

print("Mean: ({:.2f}, {:.2f})".format(mean_x, mean_y))
print("Standard Deviation:", sd)
print("Variance:", var)

# Append the first coordinate to the end of the list to close the polygon
x.append(x[0])
y.append(y[0])

# Create a new figure
fig = plt.figure()

# Add a new subplot
ax = fig.add_subplot(111)

# Plot the polygon
ax.plot(x, y, color='green', alpha=0.5)

# Set the x-axis and y-axis limits
ax.set_xlim(min(x)-1, max(x)+1)
ax.set_ylim(min(y)-1, max(y)+1)

# Set the x-axis and y-axis labels
ax.set_xlabel('X')
ax.set_ylabel('Y')

# Set the title
ax.set_title('Polygon Plot')

# Display the plot
plt.show()
```

```python
In this code, we first define the coordinates of the polygon in the x and y lists. Then we compute the mean, standard deviation, and variance of the coordinates using the mean, standard_dev, and variance functions that are modified to accept separate lists of x-coordinates and y-coordinates.

Finally, we append the first coordinate to the end of the lists to close the polygon, create a new figure using plt.figure(), add a subplot to the figure using fig.add_subplot(), plot the polygon using ax.plot(), set the x-axis and y-axis limits and labels, set the title, and display the plot using plt.show().
```
